```python
#Data cleaning 
#

df.rename(columns={
    "EmpId" : "Employment ID" , 
    "fullName" : "Full Name" ,
    "Dept" : "Department" ,
    "Age" : "Age" ,
    "Salary($)" : "Salary" ,
    "JoinDate" : "Join Date" ,
    "Performance" : "Performance" ,
    "Remote?" : "Remote"
} , inplace=True)

df
          
```

```python
#Cleaning inconsistence values on full name
df["Full Name"].to_list()
```

```python
# .strip() is used to remove spaces before and after from the strings object
# .title is used to capitalized first letter in a character
# .replace to replace a value 
# (r"[\s-]+", " ", regex=True)
# .str mean strings
df["Full Name"] = (
    df["Full Name"]
    .str.strip()
    .str.title()
    .str.replace(r"[\s-]+", " ", regex=True)
)
df["Full Name"].to_list()
```

```python
#Cleaning inconsistence values on Department
df["Department"].to_list()
```

```python
# replace 

df.replace({
    "hr" : "HR",
    "Mktng" : "Marketing",
    "marketing" : "Marketing",
    None : np.nan
}, inplace=True)

df["Department"]

```

```python
# Cleaning inconsistence value in Age
df["Age"].to_list()
```

```python
# Custom function to clean a 
# .apply() function will allow the changes on pandas

def clean_age(value): 
    if value == "thirty-five":
        return 35
    elif isinstance(value, (int, float)):
        return value
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    else:
        return np.nan

df["Age"] = df["Age"].apply(clean_age)

df["Age"]


```

```python
# assignment 
# Clean Salary, Performance and remote 
```

```python
#Cleaning inconsistence value in Salary
#df["Salary"].to_list() => display all the values in the data set 
#df["Salary"].unique() => display only the unique value in the data set 
#df["Salary"]. head() => display first 5 rows value inthe data set 
df["Salary"].to_list()
```

```python
#Cleaning inconsistence value in Salary
# .astype() to convert string value to integer/float value
df["Salary"] = (
    df["Salary"]
    .astype(str)
    .replace(r"[$,]", "", regex=True)
    .replace("NaN", np.nan)
    .astype(float)
)

df["Salary"]
```

```python
#Cleaning inconsistence value in Performance
df["Performance"].to_list()
```

```python
#Cleaning inconsistence value in Performance

df["Performance"] = (
    df["Performance"]
    .astype(str)
    .replace(r"%", "", regex=True)
    .replace("None", np.nan)
    .astype(float)
    
)

df["Performance"]
```

```python
#Cleaning inconsistence value in Remote

df["Remote"].to_list()
```

```python
#Cleaning inconsistence value in Remote 

df.replace({
        "Y" : "Yes",
        "N" : "No",
        "yes" : "Yes",
        "no" : "No",
        "None" : np.nan,
        "True" : "Yes"
    }, inplace=True)


df["Remote"]
```

```python
#cleaning inconsistence value in join_date column
# df.to_datetime()  
df["Join Date"].to_list()
```

```python
#cleaning inconsistence value in join_date column
# df.to_datetime() 
#df["Join Date"] = pd.to_datetime(df["Join Date"], errors="coerce")
#df["Join Date"]
# date utility handle any complex date format

def date_parser(date_str):
    try:
        return parser.parse(date_str, dayfirst=True)
    except Exception:
        return pd.NaT
df["Join Date"] = df["Join Date"].apply(date_parser)

df["Join Date"]
```

```python
#12/5/2026  TUESDAY
#cleaning inconsistence value in Employment ID column
# to_numeric()  => to convert a string value to a numerical value 

df["Employment ID"] = pd.to_numeric(df["Employment ID"], errors="coerce")

df["Employment ID"]
```

```python
#handling misisng values in Full Name colunm
#fillna() is used to replaced a missing value in a particular column, based on condition or alteration method
# .fillna()
df["Full Name"] = df["Full Name"].fillna("Unknown")

df["Full Name"]
```

```python
#handling misisng values in Department colunm
df["Department"] = df["Department"].fillna(df["Department"].mode()[0])

df["Department"]
```

```python
#handling misisng values in Age colunm
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Age"]
```

```python
#handling misisng values in Salary colunm
df["Salary"] = df["Salary"].fillna(df["Salary"].median())

df["Salary"]
```

```python
#handling misisng values in Performance colunm

df["Performance"] = df["Performance"].fillna(df["Performance"].median())

df["Performance"]
```

```python
#handling misisng values in Remote colunm

df["Remote"] = df["Remote"].fillna(df["Remote"].mode()[0])

df["Remote"]
```
