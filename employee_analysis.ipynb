```python
#Data cleaning 
#

df.rename(columns={
    "EmpId" : "Employment ID" , 
    "fullName" : "Full Name" ,
    "Dept" : "Department" ,
    "Age" : "Age" ,
    "Salary($)" : "Salary" ,
    "JoinDate" : "Join Date" ,
    "Performance" : "Performance" ,
    "Remote?" : "Remote"
} , inplace=True)

df
          
```

```python
#Cleaning inconsistence values on full name
df["Full Name"].to_list()
```

```python
# .strip() is used to remove spaces before and after from the strings object
# .title is used to capitalized first letter in a character
# .replace to replace a value 
# (r"[\s-]+", " ", regex=True)
# .str mean strings
df["Full Name"] = (
    df["Full Name"]
    .str.strip()
    .str.title()
    .str.replace(r"[\s-]+", " ", regex=True)
)
df["Full Name"].to_list()
```

```python
#Cleaning inconsistence values on Department
df["Department"].to_list()
```

```python
# replace 

df.replace({
    "hr" : "HR",
    "Mktng" : "Marketing",
    "marketing" : "Marketing",
    None : np.nan
}, inplace=True)

df["Department"]

```

```python
# Cleaning inconsistence value in Age
df["Age"].to_list()
```

```python
# Custom function to clean a 
# .apply() function will allow the changes on pandas

def clean_age(value): 
    if value == "thirty-five":
        return 35
    elif isinstance(value, (int, float)):
        return value
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    else:
        return np.nan

df["Age"] = df["Age"].apply(clean_age)

df["Age"]


```

```python
# assignment 
# Clean Salary, Performance and remote 
```

```python
#Cleaning inconsistence value in Salary
df["Salary"].to_list()
```

```python
#Cleaning inconsistence value in Salary
df["Salary"] = (
    df["Salary"]
    .str.replace(r"[\$,]+", "", regex=True)
)

df["Salary"]
```

```python
#Cleaning inconsistence value in Performance
df["Performance"].to_list()
```

```python
#Cleaning inconsistence value in Performance

df["Performnace"] = (
    df["Performance"]
    .str.replace(r"%", "", regex=True)

)

df["Performnace"]
```

```python
#Cleaning inconsistence value in Performance

df["Remote"].to_list()
```

```python
#Cleaning inconsistence value in Performance

df["Remote"] = (
    df["Remote"]
    .replace({
        "Y" : "Yes",
        "N" : "No",
        "yes" : "Yes",
        "no" : "No"
    } ))
df["Remote"]
```
