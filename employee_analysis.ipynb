```python
#Data cleaning 
#

df.rename(columns={
    "EmpId" : "Employment ID" , 
    "fullName" : "Full Name" ,
    "Dept" : "Department" ,
    "Age" : "Age" ,
    "Salary($)" : "Salary" ,
    "JoinDate" : "Join Date" ,
    "Performance" : "Performance" ,
    "Remote?" : "Remote"
} , inplace=True)

df
          
```

```python
#Cleaning inconsistence values on full name
df["Full Name"].to_list()
```

```python
# .strip() is used to remove spaces before and after from the strings object
# .title is used to capitalized first letter in a character
# .replace to replace a value 
# (r"[\s-]+", " ", regex=True)
# .str mean strings
df["Full Name"] = (
    df["Full Name"]
    .str.strip()
    .str.title()
    .str.replace(r"[\s-]+", " ", regex=True)
)
df["Full Name"].to_list()
```

```python
#Cleaning inconsistence values on Department
df["Department"].to_list()
```

```python
# replace 

df.replace({
    "hr" : "HR",
    "Mktng" : "Marketing",
    "marketing" : "Marketing",
    None : np.nan
}, inplace=True)

df["Department"]

```

```python
# Cleaning inconsistence value in Age
df["Age"].to_list()
```

```python
# Custom function to clean a 
# .apply() function will allow the changes on pandas

def clean_age(value): 
    if value == "thirty-five":
        return 35
    elif isinstance(value, (int, float)):
        return value
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    else:
        return np.nan

df["Age"] = df["Age"].apply(clean_age)

df["Age"]


```

```python
# assignment 
# Clean Salary, Performance and remote 
```

```python
#Cleaning inconsistence value in Salary
#df["Salary"].to_list() => display all the values in the data set 
#df["Salary"].unique() => display only the unique value in the data set 
#df["Salary"]. head() => display first 5 rows value inthe data set 
df["Salary"].to_list()
```

```python
#Cleaning inconsistence value in Salary
# .astype() to convert string value to integer/float value
df["Salary"] = (
    df["Salary"]
    .astype(str)
    .replace(r"[$,]", "", regex=True)
    .replace("NaN", np.nan)
    .astype(float)
)

df["Salary"]
```

```python
#Cleaning inconsistence value in Performance
df["Performance"].to_list()
```

```python
#Cleaning inconsistence value in Performance

df["Performance"] = (
    df["Performance"]
    .astype(str)
    .replace(r"%", "", regex=True)
    .replace("None", np.nan)
    .astype(float)
    
)

df["Performance"]
```

```python
#Cleaning inconsistence value in Remote 

df.replace({
        "Y" : "Yes",
        "N" : "No",
        "yes" : "Yes",
        "no" : "No",
        "None" : np.nan,
        "True" : "Yes"
    }, inplace=True)


df["Remote"]
```

Bar Chart
    Uses horizontal bars to compare categories.
    Best for comparing values across different groups.
    Example: Comparing sales of different products.
        
Line Chart
    Uses lines to show trends over time.
    Best for tracking changes or growth.
    Example: Monthly electricity consumption.
        
Histogram
    Shows the distribution of numerical data using grouped intervals (bins).
    Best for understanding frequency patterns.
    Example: Distribution of students’ exam scores.
        
Pie Chart
    Displays parts of a whole as slices of a circle.
    Best for showing percentages or proportions.
    Example: Market share by company.
        
Column Chart
    Similar to a bar chart but uses vertical bars.
    Best for comparing values across categories or time periods.
    Example: Yearly revenue comparison.
